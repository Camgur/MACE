```python
from pymatgen import MPRester
from pymatgen.entries.compatibility import MaterialsProjectCompatibility
from pymatgen.phasediagram.pdmaker import PhaseDiagram
from pymatgen.phasediagram.plotter import PDPlotter
from pymatgen.phasediagram.pdanalyzer import PDAnalyzer
import os
import json
from pymatgen.entries.computed_entries import ComputedEntry
%matplotlib inline

# Find the key to Materials Application Programming Interface (MAPI)
if "MAPI_KEY" not in os.environ:
    apikey = raw_input('Enter your api key (str): ')
else:
    apikey = os.environ["MAPI_KEY"]

#MAPI to extract precomputed entries from MP Database
m = MPRester(apikey)

#chemical space of interest
chemsys = ["Li", "P", "S", "Cl"]

#Get all entries from Li-P-S-Cl chemical space(mainly from MP) 
#Plus our newly ordered lowest-energy phase of Li6PS5Cl saved as entry.json
#compatibility check is performed, 
entries = m.get_entries_in_chemsys(chemsys, compatible_only=True)
entry = json.load(open('entry.json'))
new_entry=ComputedEntry.from_dict(entry)
entries.append(new_entry)
entries = MaterialsProjectCompatibility().process_entries(entries)
```

```python
#Construct the Li-P-S-Cl phase diagram
pd = PhaseDiagram(entries)

plot = PDPlotter(pd,show_unstable=False)
plt = plot.get_plot()
plt.show()
```

```python
#Generate a phase diagram analyzer
analyzer = PDAnalyzer(pd)

# There are two Li6PS5Cl polymorphs: one is in MP database (mp-985592); the other is our newly calculated one 
# Both polymorphs are metastable. Our newly calculated one is from lowest-energy ordering of disordered Li6PS5Cl in ICSD.
for e in sorted(entries, key=lambda e: e.composition.reduced_formula):
    e.ehull = analyzer.get_e_above_hull(e)
    if len(e.composition) == 4:
        print("Ehull of %s (%s) = %.1f meV/atom" % (e.composition.reduced_formula,\
                                             e.entry_id, e.ehull * 1000.0))
```
