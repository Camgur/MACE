```python
from pymatgen import Element, MPRester
from pymatgen.entries.computed_entries import ComputedEntry
from pymatgen.entries.compatibility import MaterialsProjectCompatibility
from pymatgen.phasediagram.pdmaker import PhaseDiagram
from pymatgen.phasediagram.plotter import PDPlotter
from pymatgen.phasediagram.pdanalyzer import PDAnalyzer
import os
import json

# Find the key to Materials Application Programming Interface (MAPI)
if "MAPI_KEY" not in os.environ:
    apikey = raw_input('Enter your api key (str): ')
else:
    apikey = os.environ["MAPI_KEY"]
    
#MAPI to extract precomputed entries from MP Database
m = MPRester(apikey)

#chemical space of interest
chemsys = ["Li", "P", "S", "Cl"]


#Get all entries from Li-P-S-Cl chemical space, compatibility check is performed
all_entries = m.get_entries_in_chemsys(chemsys, compatible_only=True)
print(len(all_entries))
entry = json.load(open('entry.json'))
new_entry=ComputedEntry.from_dict(entry)
all_entries.append(new_entry)
all_entries = MaterialsProjectCompatibility().process_entries(all_entries)

#Construct the phase diagram
pd = PhaseDiagram(all_entries)

#Generate a phase diagram analyzer
a = PDAnalyzer(pd)
print(len(all_entries))
```

```python
#Select newly calculated entry of Li6PS5Cl
entry=all_entries[-1]


#Get an evolution profile of Li6PS5Cl with respect to Na chemical potential
el_profile = a.get_element_profile(Element("Li"), entry.composition)

# Find the lowest Li bulk energy in eV/atom
alkali_entries = [e for e in all_entries if e.composition.reduced_formula == "Li"]
alkali_energy = min(alkali_entries, key=lambda e: e.energy_per_atom).energy_per_atom

# Print out the decomposed products of cubic Na3PS4 at varying voltage values vs Li/Li+.
# Here, voltage = -[Na chemical potential - m(Li)0] where m(Li)0 is reference chemical 
# potential of metallic Li.
for i, d in enumerate(el_profile):
    voltage = -(d["chempot"] - alkali_energy)
    print("Voltage: %s V" % voltage)
    print(d["reaction"])
    print("")
```

```python
from pymatgen.util.plotting_utils import get_publication_quality_plot
import palettable
import re
%matplotlib inline

# Plot of Li uptake per formula unit (f.u.) of Li6PS5Cl against voltage vs Li/Li+.

colors = palettable.colorbrewer.qualitative.Set1_9.mpl_colors
plt = get_publication_quality_plot(12, 8)

for i, d in enumerate(el_profile):
    v = - (d["chempot"] - alkali_energy)
    if i != 0:
        plt.plot([x2, x2], [y1, d["evolution"] / 4.0], 'k', linewidth=3)
    x1 = v
    y1 = d["evolution"] / 4.0
    if i != len(el_profile) - 1:
        x2 = - (el_profile[i + 1]["chempot"] - alkali_energy)
    else:
        x2 = 5.0
        
    if i in [0, 4, 5, 7]:
        products = [re.sub(r"(\d+)", r"$_{\1}$", p.reduced_formula if 
                           p.reduced_formula!= "Li2PS3" else "Li4P2S6") 
                    for p in d["reaction"].products if p.reduced_formula != "Li"]

        plt.annotate(", ".join(products), xy=(v + 0.05, y1 + 0.3), 
                     fontsize=24, color=colors[0])
        
        plt.plot([x1, x2], [y1, y1], color=colors[0], linewidth=5)
    else:
        plt.plot([x1, x2], [y1, y1], 'k', linewidth=3)  

plt.xlim((0, 4.0))
plt.ylim((-5, 10))
plt.xlabel("Voltage vs Li/Li$^+$ (V)")
plt.ylabel("Li uptake per f.u.")
plt.tight_layout()
```
