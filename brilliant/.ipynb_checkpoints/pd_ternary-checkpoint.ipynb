# Generate ternary phase diagram

```python
from pymatgen import MPRester
from pymatgen.entries.compatibility import MaterialsProjectCompatibility
from pymatgen.phasediagram.pdmaker import PhaseDiagram
from pymatgen.phasediagram.plotter import PDPlotter
from pymatgen.phasediagram.pdanalyzer import PDAnalyzer
import os
%matplotlib inline

# Find the key to Materials Application Programming Interface (MAPI)
if "MAPI_KEY" not in os.environ:
    apikey = raw_input('Enter your api key (str): ')
else:
    apikey = os.environ["MAPI_KEY"]

#MAPI to extract precomputed entries from MP Database
m = MPRester(apikey)
```

```python
#Li-P-S phase diagram

chemsys = ["Li", "P", "S"]
entries = m.get_entries_in_chemsys(chemsys, compatible_only=True)
entries = MaterialsProjectCompatibility().process_entries(entries)
pd = PhaseDiagram(entries)
```

```python
from matplotlib.font_manager import FontProperties
import math
import numpy as np
from pymatgen.util.string_utils import latexify
from pymatgen.util.plotting_utils import get_publication_quality_plot
from pymatgen.util.coord_utils import in_coord_list

# original phase diagram plotting object
plot = PDPlotter(pd,show_unstable=False)
plt = plot.get_plot(label_stable=False)
(lines,labels,unstable)=plot.pd_plot_data

#lines setting
for x, y in lines:
    plt.plot(x, y, "ko-", linewidth=3, markeredgecolor="k",
            markerfacecolor="b", markersize=15)

#font setting
font = FontProperties()
font.set_weight("bold")
font.set_size(24)

plt.axis("equal")
plt.xlim((-0.1, 1.2))
plt.ylim((-0.1, 1.0))
plt.axis("off")
center = (0.5, math.sqrt(3) / 6)

# Adjust the compound labels so that they are not overlapped with each other 
# nor with the lines.

for coords in sorted(labels.keys(), key=lambda x: -x[1]):
    entry = labels[coords]
    label = entry.name
    vec = (np.array(coords) - center)
    vec = vec / np.linalg.norm(vec) * 10 if np.linalg.norm(vec) != 0 \
                else vec
    valign = "bottom" if vec[1] > 0 else "top"
    if vec[0] < -0.01:
        halign = "right"
    elif vec[0] > 0.01:
        halign = "left"
    else:
        halign = "center"
#Print coords is for convenient adjusting label locations    
#    print label,coords

#Minor modification for label locations to avoide overlap and for better display    
    if label == "Li3PS4":
        coords=(0.45,0.1)
        halign = "center"
    
    if label == "P4S9":
        coords = (0.14, 0.31)
        halign = "right"
    
    if label == "P2S5":
        coords = (0.14, 0.26)
        halign = "right"
        
    plt.annotate(latexify(label), coords, xytext=vec,
                textcoords="offset points",
                horizontalalignment=halign,
                verticalalignment=valign,
                fontproperties=font)
f=plt.gcf()
f.set_size_inches((8,6))
plt.subplots_adjust(left=0.09, right=0.98, top=0.98, bottom=0.07)
plt.savefig('Li-P-S.pdf')
plt.show()
```
